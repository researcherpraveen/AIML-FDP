```python
# NumPy Solutions
import numpy as np

# Task 1: Create a 1D NumPy array containing integers from 10 to 20
arr_1d = np.arange(10, 21)
print("1D Array:", arr_1d)

# Task 2: Difference between .ndim and .shape
example_array = np.array([[1, 2, 3], [4, 5, 6]])
print("Number of dimensions (ndim):", example_array.ndim)
print("Shape of the array (shape):", example_array.shape)

# Task 3: Sum of all elements in a NumPy array
array_sum = np.sum(example_array)
print("Sum of elements:", array_sum)

# Task 4: Extract the first three elements
arr = np.array([5, 10, 15, 20])
first_three = arr[:3]
print("First three elements:", first_three)

# Task 5: Default data type and explicitly setting it to float
default_dtype = np.array([1, 2, 3]).dtype
explicit_float = np.array([1, 2, 3], dtype=float).dtype
print("Default dtype:", default_dtype)
print("Explicit float dtype:", explicit_float)

# Task 6: Element-wise addition and broadcasting
arr1 = np.array([1, 2, 3])
arr2 = np.array([[10], [20], [30]])
sum_array = arr1 + arr2  # Broadcasting
print("Element-wise addition with broadcasting:\n", sum_array)

# Task 7: Calculate mean, median, and standard deviation
stats_array = np.array([1, 2, 3, 4, 5])
print("Mean:", np.mean(stats_array))
print("Median:", np.median(stats_array))
print("Standard Deviation:", np.std(stats_array))

# Task 8: Reshape 1D array to 2D
reshaped_array = np.arange(12).reshape(3, 4)
print("Reshaped Array:\n", reshaped_array)

# Task 9: Filter elements greater than 5
arr = np.array([2, 5, 8, 1, 9])
filtered = arr[arr > 5]
print("Filtered elements:", filtered)

# Task 10: Dot product of two 2D arrays
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
dot_product = np.dot(A, B)
print("Dot product:\n", dot_product)

# Task 11: Broadcasting explained
# Example: Adding a scalar to a matrix
matrix = np.array([[1, 2], [3, 4]])
scalar = 10
broadcasted_result = matrix + scalar
print("Broadcasting example:\n", broadcasted_result)

# Task 12: Memory efficiency of NumPy
import sys
list_example = [1, 2, 3, 4, 5]
numpy_array = np.array(list_example)
print("List memory:", sys.getsizeof(list_example))
print("NumPy memory:", numpy_array.nbytes)

# Task 13: Extract even numbers without loops
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
evens = arr_2d[arr_2d % 2 == 0]
print("Even numbers:", evens)

# Task 14: Random 5x5 matrix and determinant
random_matrix = np.random.randn(5, 5)
determinant = np.linalg.det(random_matrix)
print("Random Matrix:\n", random_matrix)
print("Determinant:", determinant)
```

```python
# Pandas
import pandas as pd

# Task 1: Create a Pandas DataFrame from a dictionary of lists
data = {'Name': ['Alice', 'Bob', 'John'], 'Age': [25, 30, 35], 'sales': [200, 150, 300]}
df = pd.DataFrame(data)
print("DataFrame:\n", df)
df.to_csv('data.csv', index=False)
print("data.csv saved locally.")
# Task 2: Read a CSV file named data.csv and display the first 5 rows
df = pd.read_csv("data.csv")
print("First 5 rows:\n", df.head())

# Task 3: Find the number of rows and columns in a DataFrame
rows, columns = df.shape
print(f"Number of rows: {rows}, Number of columns: {columns}")

# Task 4: Select a single column named 'age' from a DataFrame
df_age = df['Age']
print("Age column:\n", df_age)

# Task 5: Display a summary of the DataFrame
print("DataFrame Summary:\n")
df.info()
# Task 6: Filter rows where column 'sales' has values greater than 100
filtered_df = df[df['sales'] > 100]
print("Filtered rows:\n", filtered_df)
```

```python
#pandas contd..
# Task 7: Group a DataFrame by a column named 'category' and calculate the mean of another column 'revenue'
# Sample DataFrame with 'category' and 'revenue' columns
df7 = pd.DataFrame({
    'category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'revenue': [100, 150, 200, 250, 300, 350]
})
grouped_df = df7.groupby('category')['revenue'].mean()
print("Grouped by category and mean of revenue:\n", grouped_df)

# Task 8: Replace all NaN values in a DataFrame with the median value of the respective column
# Sample DataFrame with NaN values
df8 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'John'],
    'Age': [25, None, 35],
    'sales': [200, 150, None]
})
numeric_columns = df8.select_dtypes(include=['number']).columns
df8[numeric_columns] = df8[numeric_columns].fillna(df8[numeric_columns].median())
print("DataFrame after filling NaN values with median:\n", df8)

# Task 9: Merge two DataFrames df1 and df2 on a common column 'id' using an inner join
df1 = pd.DataFrame({'id': [1, 2, 3], 'name': ['Alice', 'Bob', 'John']})
df2 = pd.DataFrame({'id': [2, 3, 4], 'revenue': [150, 200, 250]})
merged_df = pd.merge(df1, df2, on='id', how='inner')
print("Merged DataFrame on 'id' using inner join:\n", merged_df)

# Task 10: Sort a DataFrame by a column 'price' in descending order
df10 = pd.DataFrame({
    'product': ['A', 'B', 'C'],
    'price': [25, 30, 20]
})
sorted_df = df10.sort_values(by='price', ascending=False)
print("Sorted DataFrame by price (descending):\n", sorted_df)

# Task 11: Create a MultiIndex DataFrame from a dictionary where the outer index is country and the inner index is year
df11 = pd.DataFrame({
    'sales': [1000, 2000, 1500, 2500]
}, index=pd.MultiIndex.from_tuples([('USA', 2020), ('USA', 2021), ('India', 2020), ('India', 2021)],
                                   names=['country', 'year']))  # This is where the closing parenthesis is added
print("MultiIndex DataFrame:\n", df11)

# Task 12: Create a pivot table summarizing average sales for each region and product
df12 = pd.DataFrame({
    'region': ['North', 'South', 'East', 'West', 'North', 'East'],
    'product': ['A', 'B', 'C', 'A', 'B', 'C'],
    'sales': [100, 200, 150, 250, 300, 350]
})
pivot_table = df12.pivot_table(values='sales', index='region', columns='product', aggfunc='mean')
print("Pivot table summarizing average sales:\n", pivot_table)

# Task 13: Calculate the rolling mean of a column 'temperature' over a window size of 3
df13 = pd.DataFrame({
    'temperature': [20, 22, 24, 26, 28, 30]
})
df13['rolling_mean'] = df13['temperature'].rolling(window=3).mean()
print("Rolling mean of temperature:\n", df13)

# Task 14: Apply a custom function to normalize the values of a column 'score' in a DataFrame
df14 = pd.DataFrame({
    'score': [80, 85, 90, 95, 100]
})
df14['normalized_score'] = (df14['score'] - df14['score'].min()) / (df14['score'].max() - df14['score'].min())
print("DataFrame with normalized score:\n", df14)

# Task 15: Given a DataFrame with a datetime column 'date', extract the month and year as separate columns
df15 = pd.DataFrame({
    'date': pd.to_datetime(['2024-01-01', '2024-02-01', '2024-03-01'])
})
df15['month'] = df15['date'].dt.month
df15['year'] = df15['date'].dt.year
print("DataFrame with extracted month and year:\n", df15)
```

```python
#Data visualization
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 1. Plot a simple line graph with x = [1, 2, 3, 4] and y = [10, 20, 25, 30]
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.show()

# 2. Add a title, x-axis label, and y-axis label to a plot in Matplotlib
plt.plot(x, y)
plt.title('Simple Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# 3. Create a bar plot using Seaborn for the dataset df with x='category' and y='values'
# Sample data
df = pd.DataFrame({
    'category': ['A', 'B', 'C', 'D'],
    'values': [5, 10, 15, 20]
})
sns.barplot(x='category', y='values', data=df)
plt.show()

# 4. Load the tips dataset in Seaborn and display the first 5 rows
tips = sns.load_dataset('tips')
print(tips.head())

# 5. What function is used to display the plot in Matplotlib or Seaborn?
# The function used is `plt.show()` in Matplotlib, or `sns.show()` in Seaborn. This function displays the plot.
# Example:
plt.plot(x, y)
plt.show()

# 6. Create a figure with 2 rows and 1 column of subplots, showing two different line graphs
fig, axes = plt.subplots(2, 1)  # 2 rows, 1 column
axes[0].plot(x, y)
axes[0].set_title('Line Graph 1')
axes[1].plot([1, 2, 3, 4], [5, 10, 15, 20])
axes[1].set_title('Line Graph 2')
plt.tight_layout()  # Adjust layout
plt.show()

# 7. Add a legend to a plot in Matplotlib and specify its location (e.g., top right)
plt.plot(x, y, label='Line 1')
plt.legend(loc='upper right')  # Location of the legend
plt.show()

# 8. Create a scatter plot using Seaborn with points colored by a categorical column, and add a regression line
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=["o", "s"], aspect=1.5)
plt.show()

# 9. Pairplot in Seaborn, create one for numerical columns in the tips dataset
sns.pairplot(tips.select_dtypes(include=['float64', 'int64']))  # Pairplot for numerical columns only
plt.show()

# 10. Create a boxplot in Seaborn, customizing its color palette to coolwarm
sns.boxplot(x='day', y='total_bill', data=tips, palette='coolwarm')
plt.show()

# 11. Create a dual-axis plot (e.g., line plot on the left y-axis and a bar plot on the right y-axis)
fig, ax1 = plt.subplots()

# Line plot on the left y-axis
ax1.plot(x, y, color='tab:blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Line Plot', color='tab:blue')

# Create a second y-axis
ax2 = ax1.twinx()
ax2.bar(x, y, color='tab:orange')
ax2.set_ylabel('Bar Plot', color='tab:orange')

plt.show()

# 12. Modify the tick labels on both x and y axes to have a specific font size and rotation
plt.plot(x, y)
plt.xticks(fontsize=12, rotation=45)
plt.yticks(fontsize=12, rotation=90)
plt.show()

# 13. Create a heatmap in Seaborn with annotated values, customizing the color map to viridis
data = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
sns.heatmap(data, annot=True, cmap='viridis')
plt.show()

# 14. Create a FacetGrid in Seaborn, splitting the dataset by a categorical column and plotting histograms for each group
g = sns.FacetGrid(tips, col='sex')  # Split by 'sex' column
g.map(sns.histplot, 'total_bill')
plt.show()

# 15. Create a violin plot using Seaborn with split violins for a categorical variable and overlay a swarm plot for individual data points
sns.violinplot(x='day', y='total_bill', data=tips, split=True, inner='quart')
sns.swarmplot(x='day', y='total_bill', data=tips, color='k', alpha=0.5)
plt.show()
```
